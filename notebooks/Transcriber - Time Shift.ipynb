# Script base

```python
fn_srt_chinese_original = 'Play Digital Signal Processing with Python (Chinese).srt'
srt_chinese = SRT(fn_srt_chinese_original)
srt_chinese.scripts
```

# Script to be shifted in time

```python
fn_srt_english_original = 'Play Digital Signal Processing with Python (English) - original.srt'
srt_english = SRT(fn_srt_english_original)
srt_english.scripts
```

# Time Delta

```python
s1_base = srt_chinese.get_script(1)
print(s1_base)
```

```python
s1_to_be_shifted = srt_english.get_script(1)
print(s1_to_be_shifted)
```

```python
td = s1_to_be_shifted.start_time - s1_base.start_time
td
```

```python
srt_english.time_shift(-td)
srt_english.scripts[:3]
```

# Captions

```python
fn_srt = 'Play Digital Signal Processing with Python (English).srt'
print(srt_english.dump(fn_srt))
```
